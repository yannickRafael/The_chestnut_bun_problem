```python
import matplotlib.pyplot as plt
```

```python
def calculate_growth_time(initial_volume, target_volume, growth_rate):
    time = 0
    current_volume = initial_volume

    while current_volume < target_volume:
        current_volume *= growth_rate
        time += 5

    return time
```

```python
initial_chestnut_volume = int(input("insert the initial volume of the chestnut"))
solar_system_volume = int(input("insert the1 volume of the solar system"))
growth_rate = int(input("insert the growth rate"))
```

```python
time_to_cover_solar_system = calculate_growth_time(initial_chestnut_volume, solar_system_volume, growth_rate)
```

```python
print(f"Time to cover the solar system with chestnut buns: {time_to_cover_solar_system} minutes")
```

```python
time_points = list(range(0, time_to_cover_solar_system + 5, 5))
volume_points = [initial_chestnut_volume * (growth_rate**(t/5)) for t in time_points]
```

```python
plt.plot(time_points, volume_points)
plt.xlabel('Time (minutes)')
plt.ylabel('Volume of Chestnut Buns')
plt.title('Chestnut Bun Growth Over Time')
plt.show()
```
