```python
import matplotlib.pyplot as plt
import numpy as npy
```

```python
def calculate_amount(time):
    time_passed = 0
    amount = 1
    amounts = []
    timeline = []
    while time >= 5:
        amounts.append(amount)
        timeline.append(time_passed)
        amount = amount * 2
        time_passed += 5
        time -= 5
    amounts.append(amount)
    timeline.append(time_passed)
    return amounts, timeline


```

```python
time = int(input('insert the time passed after the sprinkle(in minutes)'))
amounts, timeline = calculate_amount(time)
```

```python
x = npy.array(timeline)
y = npy.array(amounts)
print(x)
```

```python
plt.title('Title')
plt.xlabel('time')
plt.ylabel('amount')
```

```python
plt.plot(x, y, color='black', marker='o', linewidth=2, markersize=5, markerfacecolor='blue')
plt.grid(True)
```
